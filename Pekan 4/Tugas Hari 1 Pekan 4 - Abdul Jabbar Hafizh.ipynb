<h1>Jawaban Soal 1: Pemahaman Machine Learning</h1>

Jawab/Kerjakan pertanyaan/perintah di bawah ini dengan bahasa kalian sendiri:

- Apa itu machine learning?
Machine Learning adalah kemampuan mesin untuk mengakses data yang ada dengan perintah mereka sendiri. Machine Learning juga mampu mempelajari data yang ada dan melakukan tugas-tugas tertentu. Machine Learning mampu melakukan ini dengan metode mempelajari algoritma dan model statistik yang ada.

- Apa itu data feature dan data target?

Feature adalah "variabel" atau "atribut" dalam kumpulan data (dataset). Bergantung pada apa yang Anda coba analisis, fitur yang Anda sertakan dalam kumpulan data Anda dapat sangat bervariasi.


Target adalah variabel yang nilainya akan dimodelkan dan diprediksi oleh variabel lain.

- Apa Perbedaan Supervised Learning dan Unsupervised Learning?
Supervised learning membutuhkan data training (harus dilatih terlebih dahulu) sedangkan unsupervised learning tidak membutuhkan data training (tidak perlu dilatih terlebih dahulu).

- Apa Jenis2 yang ada di dalam Supervised Learning? Jelaskan Perbedaannya!
Clustering: Masalah clustering adalah saat Anda ingin menemukan pengelompokan yang melekat dalam data, seperti mengelompokkan pelanggan menurut perilaku pembelian.

Associating: Masalah associating adalah saat Anda ingin menemukan aturan yang menjelaskan sebagian besar data Anda, seperti orang yang membeli X dan juga cenderung membeli Y.



- Apa perbedaan Hyperparameter dan Parameter?
Hyperparameter adalah variabel yang memengaruhi output model. Dengan kata lain, nilai hyperparameter tidak bergantung pada data dan selalu kita ambil as given saat pendefinisian model

Sedangkan parameter adalah variabel yang tidak selalu dimiliki oleh sebuah class. Adanya parameter mampu membuat sebuah fungsi memetakan lebih banyak data dari domain input ke domain output dengan tepat.

Bedanya, nilai hyperparameter tidak diubah selama model dioptimisasi. 

- Sebutkan Langkah-langkah dalam mengaplikasikan algoritma apapun dalam machine learning!
    1. Definisi masalah.
    2. Menyiapkan data.
    3. Memilih/mengevaluasi algoritma.
    4. Melakukan tuning untuk memperbaiki hasil.
    5. Menyajikan hasil.

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
model_lr = DecisionTreeRegressor(max_depth=3)
x_matrix = x[:, np.newaxis]
model_lr.fit(x_matrix, y)
x_test = np.linspace(20, 30, 50)
x_test = x_test[:, np.newaxis]
y_test = model_lr.predict(x_test)
y_train = model_lr.predict(x_matrix)

fig, ax = plt.subplots(figsize=(16, 8))
ax.scatter(x, y, label='Data asli')
ax.scatter(x, y_train, color='y', label='Prediksi terhadap data training')
ax.plot(x_test, y_test, color='r', label='Prediksi terhadap data baru')
ax.legend(loc='upper left', bbox_to_anchor=(0.9, 1.2))
plt.show()
```
