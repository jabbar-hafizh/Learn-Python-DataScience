<h2>Soal 2: Aritmatika di Python </h2>
    
- Buat statement pertambahan antara 2 number di Python
- Buat statement perkurangan antara 2 number di Python
- Buat statement perkalian antara 2 number di Python
- Buat statement pembagian antara 2 number di Python

```python
a = input("Input number of a: ")
b = input("Input number of b: ")

# Statement operasi aritmatika dengan input
c=int(a)+int(b)
print('Statement Pertambahan:',a, '+' ,b, '=',c)

c=int(a)-int(b)
print('Statement Pengurangan:',a, '+' ,b, '=',c)

c=int(a)*int(b)
print('Statement Perkalian:',a, '+' ,b, '=',c)

c=int(a)/int(b)
print('Statement Pembagian:',a, '+' ,b, '=',c)
```
