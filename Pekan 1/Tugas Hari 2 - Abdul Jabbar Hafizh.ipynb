<h2>Soal 2: Built-in Function</h2>

Jawab Pertanyaan di bawah ini:
    
Apa itu built-in Function?

Sebutkan 3 built-in Function di pyhon!

Berikan contoh masing2 1 cara penggunaan built-in function yang kamu sebutkan sebelumnya

Built-in function adalah function bawaan yang ada pada python.
---
3 contohnya:

1.   Built-in len(), mengembalikan panjang objek.
2.   Built-in sum(), menjumlahkan semua bilangan atau item yang menjadi argumennya.
3.   Built-in len(), mengembalikan list yang disortir.



```python
# Contoh penggunaan built in function
bilangan = [3, 1, 5, 7]

# Mengembalikan panjang objek
print(len(bilangan))

# Menjumlahkan semua bilangan atau item yang menjadi argumennya.
print(sum(bilangan))

# Mengembalikan list yang disortir
print(sorted(bilangan))
```

<h2>Soal 6: Membuat Function dengan Multiple arguments</h2>

Buatlah suatu fungsi untuk melakukan penggabungan antara dua list

```python
obj_list = [2, 4, 5, 6]
obj_penambah = [1, 2, 3]

def kali_list(obj_list, obj_penambah):
  combine = obj_list + obj_penambah
  return combine
    
    
print(kali_list(obj_list, obj_penambah))
```

Expected Output:

[2, 4, 5, 6, 1, 2, 3]
