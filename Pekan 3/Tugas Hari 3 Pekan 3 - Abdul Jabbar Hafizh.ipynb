```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(figsize=(18,8), nrows=1, ncols=3, sharey=True)
# plt.style.use('default')

ax[0].plot(x, tan, color="blue", label="tanX")
ax[1].plot(x, sin, color="red", label="sinX")
ax[2].plot(x, cos, color="green", label="cosX")

ax[0].set_title("Grafik tanx")
ax[1].set_title("Grafik sinx")
ax[2].set_title("Grafik cosx")
fig.legend(loc=9)
# ax.legend()
fig.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(18,8), sharex=True)

ax.plot(x,y, color='b')
ax.set_title('Grafik CosX')
ins2 = ax.inset_axes(bounds=[-0.1,-0.1,0.25,0.3])
ins2.plot(x,y2, color='green')
ins2.plot(x,y)
ins2.set_title('Grafik CosX^2')
ins3 = ax.inset_axes(bounds=[0.9,-0.1,0.25,0.3])
ins3.plot(x,y3, color='red')
ins3.plot(x,y)
ins3.set_title('Grafik CosX^3')
ins4 = ax.inset_axes(bounds=[0.9,0.9,0.25,0.3])
ins4.plot(x,y4, color='black')
ins4.plot(x,y)
ins4.set_title('Grafik CosX^4')
ins5 = ax.inset_axes(bounds=[-0.1,0.9,0.25,0.3])
ins5.plot(x,y5, color='orange')
ins5.plot(x,y)
ins5.set_title('Grafik CosX^5')

fig.show()
```
