<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)

# membuat objek figure
fig = plt.figure()
# membuat objek axes
ax = fig.add_subplot()
# menampilkan objek visual
plt.show()

# cara explicit
# menambahkan dalam ke objek axes
ax.plot(x, y)
# menampilkan data
plt.show()

# cara implicit
# membuat objek visual
plt.plot(x, y)
# menampilkan objek visual
plt.show()
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
from IPython.display import Image

plt.style.use('seaborn')
fig, a = plt.subplots()
a.plot(x,y)
a.plot(x,y2)
a.plot(x,y3)
plt.show()

# menyimpan file ke files
fig.savefig('a.png')
# menampilkan file dari files
Image('a.png')
```
