Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.style.use('default')

ax.plot(x, cos, color="green", label="Cosx")
ax.plot(x, cos3, linestyle='--', color="red", label="3 Cosx")
ax.plot(x, sin, linestyle='-.', color="blue", label="Sin(x-90)")
ax.plot(x, sin2, linestyle='', marker='x', color="yellow", label="2 Sinx")

# mengubah axis label dan title
ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')
ax.legend(loc=10)
plt.show()
```
