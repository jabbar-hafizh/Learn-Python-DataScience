<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
> Series List
- index seriesnya tidak harus bertipe int
- Ketika print obj_series akan menampilkan index dan value nya.
- Ketika print list akan menampilkan value nya saja.
> Series Dict
- Dapat mempunyai duplicate key yang sama
- Bisa slicing

- Apa perbedaan dataframe dan series

> Perbedaan dari Series dan DataFrame ini terletak pada strukturnya dan juga attribute dan method-method yang mereka miliki, untuk perbedaan strukturnya adalah sebagai berikut :


*   > Series adalah objek 1 dimensi. Series adalah suatu objek satu dimensi yang dapat menyimpan berbagai jenis tipe data seperti integer, string, dan lain sebagainya. tipe data dari objek series ini harus seragam.

*   > DataFrame adalah suatu objek 2 dimensi, mirip seperti tabel. DataFrame adalah suatu objek 2 dimensi tempat menyimpan data dengan lebih terstruktur. dataframe memiliki 2 index, yaitu index baris dan index columns. Dalam satu column dataframe harus memiliki tipe data yang sama. tapi antar columnnya dataframe bisa memiliki jenis data yang berbeda.




- Buatlah suatu data series

Jawab: 

```python
# code here
# cara pertama
x = pd.Series([6,3,4,6])
print(x)

x.index = ['a', 'b', 'c', 'd']
print(x)

# cara kedua
x = pd.Series([6,3,4,6], index=['a', 'b', 'c', 'd'])
print(x)
print(type(x))
```

<h2>Soal 5: Deskriptif Statistik</h2>

Jelaskan apa itu deskriptis statistik!

Jawab: 
> Statistika deskriptif merupakan suatu metode yang berhubungan dengan pengumpulan atau penyajian data sampai memberi informasi yang berguna.

Statistika di golongkan menjadi dua bagian, yakni statistika deskriptif dan statistika inferensia yang dilakukan berdasarkan aktivitas yang dilakukan.

Untuk contoh dari statistika deskriptif sendiri antara lain yaitu:

- tabel
- diagram
- grafik
- besaran-besaran lain dalam majalah dan koran-koran.

> Statistika deskriptif berkaitan dengan bagaimana kita
menjelaskan dan murumuskan suatu data. Umumnya digunakan dua metode pendekatan:

- Pendekatan kuantitatif menjelaskan data secara
numerik.
- Pendekatan visual mengilustrasikan data
menggunakan chart, plot, histogram, atau grafik lainnya.
Kita dapat menerapkan statistika deskriptif ke dalam satu atau banyak data. Ketika kita menjelaskan dan merumuskan suatu data tunggal, maka kita sedang melakukan analisa univariate. Ketika kita mencari hubungan statistika antara sepasang data, maka kita sedang melakukan analisa bivariate. Begitu juga dengan analisa multivariate yang berarti kita menerapkannya pada banyak data sekaligus.



<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# code here
# menentukan dataframe
c_price = df_diamonds['price']
c_depth = df_diamonds['depth']
c_carat = df_diamonds['carat']

# rata-rata column price
print(f'Rata-rata Column Price: {c_price.mean()}\n')

# standar deviasi column depth
print(f'Standar Deviasi Column Depth: {c_depth.std()}\n')

# max dan min column carat
print('Max dan Min dari Carat\n''Max: ',c_carat.max(),'\nMin: ',c_carat.min())
```
